```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


from sklearn import tree 
from sklearn.tree import DecisionTreeClassifier
```

```python
X = np.load('x.npy')
y = np.load('y.npy')
```

## optimal hyperparameter's

```python
def hyperParameters_plot(max_depth, max_features, all_acc_scores):
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(projection='3d')
    
    D, F = np.meshgrid(range(1, max_depth+1), range(1, max_features+1))
    
    # Transpose D and F
    D, F = D.T, F.T
    
    ax.plot_surface(D, F, np.array(all_acc_scores), cmap='viridis', edgecolor='k')
    
    ax.set_xlabel('Max Depth')
    ax.set_ylabel('Max Features')
    ax.set_zlabel('Accuracy')
    
    plt.show()
```

```python
def optimal_hyperParameters(X, y, max_depth, max_features, plot=False):
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1585)
    all_acc_scores = []
    optimal_accuracy = 0
    optimal_depth, optimal_feature = 0, 0
    
    for d in range(1, max_depth+1):
        acc_scores = []
        
        for f in range(1, max_features+1):
            clf = DecisionTreeClassifier(max_depth=d, max_features=f)
            clf = clf.fit(X_train, y_train)
            y_pred = clf.predict(X_test)
            acc = round(accuracy_score(y_test, y_pred) * 100, 2)
            acc_scores.append(acc)
            
            if acc > optimal_accuracy:
                optimal_accuracy = acc
                optimal_depth, optimal_feature = d, f
        
        all_acc_scores.append(acc_scores)
    
        
    if plot:
        hyperParameters_plot(max_depth, max_features, all_acc_scores)
        
    return optimal_depth, optimal_feature, optimal_accuracy
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1585)

max_depth = 10
max_feature = int(0.8 * X.shape[1])
```

```python
# Training set
optimal_depth_train, optimal_feature_train, optimal_accuracy_train = optimal_hyperParameters(
    X_train, y_train, max_depth, max_feature, plot=True
)
```

```python

# Test set
optimal_depth_test, optimal_feature_test, optimal_accuracy_test = optimal_hyperParameters(
    X_test, y_test, max_depth, max_feature, plot=False
)
```

```python

print("Training Set Results:")
print("Optimal Depth:", optimal_depth_train)
print("Optimal Max Features:", optimal_feature_train)
print("Optimal Accuracy:", optimal_accuracy_train)

print("\nTest Set Results:")
print("Optimal Depth:", optimal_depth_test)
print("Optimal Max Features:", optimal_feature_test)
print("Optimal Accuracy:", optimal_accuracy_test)
```

## plot learning curve

```python
from sklearn.model_selection import learning_curve
```

```python
def plot_learning_curve(classifier, X, y, title):
    train_sizes, train_scores, test_scores = learning_curve(
        classifier, X, y, train_sizes=np.linspace(0.25, 1.0, num=4), cv=5
    )
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.figure(figsize=(10, 6))
    plt.title(title)
    plt.ylabel('Score')
    plt.xlabel('Training Examples')

    plt.grid()
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std, train_scores_mean + train_scores_std, alpha=0.1, color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std, test_scores_mean + test_scores_std, alpha=0.1, color="b")

    plt.plot(train_sizes, train_scores_mean, 'o-', label="Training Score", color="r")
    plt.plot(train_sizes, test_scores_mean, 'o-', label="Testing Score", color="b")

    plt.legend()
    plt.show()
```
